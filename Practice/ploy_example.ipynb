```python
import numpy as np 
import pandas as pd
```

# load data

```python
l1 = [i for i in range(1,20) if i%2 == 0]

l2 = [i**3 for i in range(1,10)]

df = pd.DataFrame({'X':l1, 'y':l2})

print(l1)
print(l2)
```

# defferentiate Dep. or indep.

```python
X = df.iloc[:,:1]
y = df.iloc[:,1]

print(type(X))
print(type(y))
```

```python
import seaborn as sns

sns.scatterplot(x = df.X , y = df.y, data = df)
```

# modeling

```python

from sklearn.linear_model import LinearRegression

regressor = LinearRegression()

model = regressor.fit(X, y)

print(model.coef_)
print(model.intercept_)
```

# Evalution

```python
df['ypred'] = model.predict(X)

from sklearn.metrics import mean_squared_error, r2_score

MSE = mean_squared_error(y, df.ypred)
RMSE = np.sqrt(MSE)
R_Squared = r2_score(y , df.ypred)

print('Root Mean Squared Error : ' , RMSE)
print('R Squared : ', R_Squared)
```

```python
sns.scatterplot(x = df.X , y = df.y , data = df)
sns.lineplot(x = df.X, y = df.ypred , data = df, color = 'red')
```

```python
df['X2'] = df.X**3

model_2 = regressor.fit(df[['X','X2']],y)

df['ypred_2'] = model_2.predict(df[['X','X2']])
```

```python
sns.scatterplot(x = df.X , y = df.y , data = df)
sns.lineplot(x = df.X, y = df.ypred_2 , data = df, color = "red")
```

```python
MSE_2 = mean_squared_error(y, df.ypred_2)
RMSE_2 = np.sqrt(MSE)
R_Squared_2 = r2_score(y , df.ypred_2)

print('Root Mean Squared Error : ' , RMSE_2)
print('R Squared : ', R_Squared_2)
```
